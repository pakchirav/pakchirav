## LivingInsider Web Scraping
 Extracting rental condo data in Samyan area



```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

buildings = []
sq_m = []
floors = []
n_bedrooms = []
n_restrooms = []
full_prices = []
prices = []
views = []
texts = []
clean_texts = []

for i in range(0, 28) :
    url = "https://www.livinginsider.com/living_zone/5/Condo/Rent/" + str(i+1) + "/%E0%B8%AA%E0%B8%A2%E0%B8%B2%E0%B8%A1-%E0%B8%88%E0%B8%B8%E0%B8%AC%E0%B8%B2-%E0%B8%AA%E0%B8%B2%E0%B8%A1%E0%B8%A2%E0%B9%88%E0%B8%B2%E0%B8%99.html"
    r = requests.get(url)
    bs = BeautifulSoup(r.text,"lxml")
    #print(bs)
    #print(bs.prettify())
    #print(bs.div)
    elements = bs.find_all('div', {'class': 'item-desc'})
    #print(len(elements))
    for element in elements:
        building = element.find('div', {'class' : 'col-md-12 col-sm-12 col-xs-12'}).text.strip()
        size = element.find('div', {'class' : 'max_width_show_data_building'}).text.strip()
        size = size.split(' ')[0]
        floor_tag = element.find('div', {'class': 'col-md-6 col-sm-6 col-xs-6 ic-detail'})
        floor = floor_tag.text.strip().split(' ')[-1]
        bedroom = floor_tag.find_next_sibling('div').text.strip()
        bedroom = bedroom.split(' ')[0]
        restroom = element.find('div', {'class' : 'col-xs-6 ic-detail col-md-6 col-sm-6'}).text.strip()
        restroom = restroom.split(' ')[0]
        price_element = element.find('div', {'class' : 'listing_cost'})
        if price_element :
            text = price_element.text.strip()
            clean_text = text.replace('\n', '')
            price_part = clean_text.split(' ')
            #print(price_part)
            if len(price_part) > 1:
                full_price = price_part[1]
                price = price_part[-1][1:]
            else :
                full_price = None
                price = price_part[0][1:]
                #print(full_price)
                #print(price)
        view = element.find('div', {'class' : 'istock-view pull-right'}).text.strip()

        buildings.append(building)
        sq_m.append(size)
        floors.append(floor)
        n_bedrooms.append(bedroom)
        n_restrooms.append(restroom)
        texts.append(text)
        full_prices.append(full_price)
        prices.append(price)
        views.append(view)
        clean_texts.append(clean_text)

#print(buildings)
#print(sq_m)
#print(floors)
#print(n_bedrooms)
#print(n_restrooms)
#print(full_prices)
#print(prices)
#print(views)
#print(texts)
#print(clean_texts)

data = {'Building' : buildings, 'Size' : sq_m, 'Floor' : floors, 'Bedroom' : n_bedrooms,
            'Restroom' : n_restrooms, 'FullPrice' : full_prices, 'Price' : prices, 'View' : views}
df = pd.DataFrame(data)
#print(df)
```

```python
df.to_csv('rentalcondo_scraping.csv', index=False, encoding = 'utf-8-sig')
```
